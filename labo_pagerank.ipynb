Dans ce laboratoire, vous aurez besoin de 4 notions en Python :


1.   Définir des matrices
2.   Élever une matrice à une puissance

Voici des exemples détaillés de ces 3 notions.



## 1. Définir des matrices

*   Vous devez d'abord **importer la librairie numpy**. Nous allons nommer cette librairie np pour simplifier la lecture du code. *Cette action doit être fait une seule fois dans tout le laboratoire.*

* Vous devez ensuite créer votre matrice à l'aide de la commande `np.array`.



Les éléments de la matrice doivent être rentrés ligne par ligne :

*   Tous les éléments d'une ligne doivent se trouver entre crochets "[ ]" et doivent être séparés par une virgule.
*   Pour changer de ligne, il suffit de fermer votre crochet, de séparer à l'aide d'une virgule et d'ouvrir un nouveau crochet pour la ligne suivante.

Voici un exemple :

```python
# Cette ligne sert à importer la librairie numpy :
import numpy as np
from numpy import linalg

# Je définie une matrice nommée "M_exemple". Vous pouvez donner le nom que vous
# souhaitez à votre matrice. Remarquez que la matrice complète est entre
# crochets.
M_exemple1 = np.array([
    [11, 12, 13],
    [21, 22, 23],
    [31, 32, 33]
])
M_exemple2 = np.array([
    [1],
    [2],
    [3]
])

# Cette ligne permet d'afficher la matrice "M_exemple"
print("Voici la matrice 1 : \n", M_exemple1)
print("Voici la matrice 2 : \n", M_exemple2)
```

## 2. Élever une matrice à une puissance

* Vous devez utiliser la commande `np.linalg.matrix_power(votre_matrice,votre_puissance)`.

Voici un exemple :

```python
M_puissance5 = np.linalg.matrix_power(M_exemple1, 5)
print(M_puissance5)
```

## 3. Produit de matrices

* Vous devez séparer les deux matrices à multiplier avec l'opérateur `@`.

Voici un exemple :

```python
produit = M_exemple1 @ M_exemple2
print(produit)
```

## 4. Calculer un déterminant de matrice


*   Vous devez utiliser la commande `np.linalg.det(votre_matrice)`





```python
determinant = np.linalg.det(M_exemple1)
print(determinant)
```

### **Question 1 f)**
Complétez le code suivant pour définir vos matrices $T$ et $P_0$

```python
import numpy as np

T = np.array([
    [0, 0, 0, 1 / 3, 1],
    [1 / 4, 0, 1 / 2, 1 / 3, 0],
    [1 / 4, 1 / 2, 0, 0, 0],
    [1 / 4, 1 / 2, 1 / 2, 0, 0],
    [1 / 4, 0, 0, 1 / 3, 0]
])

print(T)
print(T.shape)

P_0 = np.array([
    [0],
    [1],
    [0],
    [0],
    [0]
])

print(P_0)
print(P_0.shape)
```

### **Question 1 g)**
Définissez votre matrice $P_2$ dans l'environnement.

```python
P_2 = T @ T @ P_0
print(P_2)
print(P_2.shape)
```
